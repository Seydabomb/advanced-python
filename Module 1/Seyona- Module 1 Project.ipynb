```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
# Question 1- creating a scatter plot that shows land size on the x-axis and price on the y-axis. Change the color of the points to reflect the type of the property. Paste the image below.
df.sort_values(by='Landsize',inplace=True)
plt.figure(figsize=(16,8))
sns.scatterplot(x='Landsize',y='Price',data=df, hue='Type')

plt.show()
```

```python
# Question 3- creating a line chart that shows the year built along the x-axis and the average price per year on the y-axis.
df.sort_values(by='YearBuilt',inplace=True)
avg_price_per_year = df.groupby('YearBuilt')['Price'].mean().reset_index()

plt.figure(figsize=(16,8))
sns.lineplot(x='YearBuilt', y='Price', data=avg_price_per_year)
plt.show()
```

```python
# Question 5- creating a bar chart that shows the region on the x-axis and the average price of the property on the y-axis. 
df.sort_values(by='Regionname',inplace=True)
avg_price_of_property = df.groupby('Regionname')['Price'].mean().reset_index()

plt.figure(figsize=(16,8))
sns.barplot(x='Regionname',y='Price',data=avg_price_of_property)

plt.xlabel('Region Name')
plt.ylabel('Average Price of Property')
plt.title('Average Price of Property By Region')
plt.xticks(rotation=30,ha='center')
plt.show()
```

```python
# Question 7- creating a histogram of property prices. 

plt.figure(figsize=(16,8))
sns.histplot(x='Price',data=df)
plt.show()
```

```python
# Question 9- creating a box plot that shows the years that each property was built on the x-axis and the region on the y-axis. These box plots should be oriented horizontally. 
plt.figure(figsize=(16,8))
sns.boxplot(x='YearBuilt',y='Regionname', data=df)

plt.xlabel('Year Built')
plt.ylabel('Region Name')
plt.title('Years Each Property Was Built By Region')
plt.show()
```
