```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
# Question 1- creating a countplot that shows a count of how many of the companies in the data set are from each state
df.sort_values(by='state',inplace=True)
plt.figure(figsize=(16,8))
sns.countplot(x='state',data=df)
plt.xticks(rotation=45,ha='center')
plt.show()
```

```python
# Question 3- creating a bar chart that shows a sum of how many total job openings there are in each industry.
df.sort_values(by='Industry',inplace=True)
plt.figure(figsize=(16,8))
sns.countplot(x='Industry',data=df)
plt.xticks(rotation=30,ha='center')
plt.show()
```

```python
# Question 5- creating a line chart that shows the average growth rate for each founded year.
df.sort_values(by='founded',inplace=True)
avg_growth_rate = df.groupby('founded')['growth_percentage'].mean().reset_index()

plt.figure(figsize=(16,8))
sns.lineplot(x='founded',y='growth_percentage',data=avg_growth_rate)
plt.xticks(rotation=30,ha='center')
plt.show()
```

```python
# Question 7- creating a scatter plot that compares the number of employees to estimated revenues for each company.

df.sort_values(by='employees',inplace=True)
plt.figure(figsize=(16,8))
sns.scatterplot(x='employees',y='estimated_revenues',data=df)
plt.xticks(rotation=30,ha='center')
plt.show()
```

```python
# Question 9- creating a histogram that measures the number of companies founded in each year.
plt.figure(figsize=(16,8))
sns.histplot(x='founded',data=df)

plt.xlabel('Year Founded')
plt.ylabel('Number of Companies')
plt.title('Number of Companies Founded Each Year')
plt.show()
```
